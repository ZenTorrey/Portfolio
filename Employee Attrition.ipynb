## Dealing with Categorical Data

```python
dataset = pd.get_dummies(data=dataset, drop_first=True)
```

## Splitting The Dataset

```python
x = dataset.drop(columns='Attrition')
```

```python
y = dataset['Attrition']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)
```

```python
x_train.shape
```

```python
y_train.shape
```

```python
x_test.shape
```

```python
y_test.shape
```

## Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```python
x_train
```

```python
x_test
```

# Building The Model

## Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
classifer_lr = LogisticRegression(random_state=0)
classifer_lr.fit(x_train, y_train)
```

```python
y_pred = classifer_lr.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
acc = accuracy_score(y_test, y_pred)
print(acc*100)
```

```python
cm = confusion_matrix(y_test, y_pred)
print(cm)
```

## Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
classifer_rf = RandomForestClassifier(random_state=0)
classifer_rf.fit(x_train, y_train)
```

```python
y_pred = classifer_rf.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix
acc = accuracy_score(y_test, y_pred)
print(acc*100)
```

```python
cm = confusion_matrix(y_test, y_pred)
print(cm)
```

## Support Vector Machine

```python
from sklearn.svm import SVC
classifer_svc = SVC(random_state=0)
classifer_svc.fit(x_train, y_train)
```

```python
y_pred = classifer_svc.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix
acc = accuracy_score(y_test, y_pred)
print(acc*100)
```

```python
cm = confusion_matrix(y_test, y_pred)
print(cm)
```

# Randomized Search to Find the Best Parameters (Logistic Regression)

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
parameters = {
    'penalty': ['l2'],  # Only l2 penalty for lbfgs and newton-cg solvers
    'C': [0.25, 0.5, 0.75, 1.0, 1.25, 1.5, 175, 2.0],
    'solver': ['newton-cg', 'lbfgs'],  # Use only lbfgs and newton-cg solvers
    'max_iter': [50, 100, 500, 2000, 5000]
}

classifier_lr = LogisticRegression(random_state=0)
random_cv = RandomizedSearchCV(estimator=classifier_lr, param_distributions=parameters, 
                               n_iter=10, scoring='roc_auc', n_jobs=-1, cv=5, verbose=3)
random_cv.fit(x_train, y_train)
```

```python
random_cv.best_estimator_
```

```python
random_cv.best_params_
```

```python
random_cv.best_score_
```

# Final Model (Logistic Regression)

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
sc.fit(x_train)
x_train_scaled = sc.transform(x_train)
x_test_scaled = sc.transform(x_test)
classifier = LogisticRegression(C=0.5, max_iter=2000, random_state=0, solver='sag')
classifier.fit(x_train_scaled, y_train)
```

```python
y_pred = classifier.predict(x_test_scaled)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix
acc = accuracy_score(y_test, y_pred)
print("Accuracy:", acc * 100)
```

```python
cm = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:")
print(cm)
```
